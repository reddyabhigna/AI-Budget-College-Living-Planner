```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Student Profile
income = float(input("Enter Monthly Income (₹): "))
living_type = input("Living Type (Hostel / PG / Home): ")

rent = float(input("Rent / Hostel Fees (₹): "))
food = float(input("Food Expense (₹): "))
travel = float(input("Travel Expense (₹): "))
entertainment = float(input("Entertainment Expense (₹): "))
others = float(input("Other Expenses (₹): "))
```

```python
total_expense = rent + food + travel + entertainment + others
savings = income - total_expense

fixed = rent
variable = food + travel + entertainment + others

print("\n----- Budget Summary -----")
print("Total Expense:", total_expense)
print("Remaining Savings:", savings)
```

```python
categories = ['Fixed', 'Variable']
amounts = [fixed, variable]

plt.bar(categories, amounts)
plt.title("Expense Categorization")
plt.ylabel("Amount (₹)")
plt.show()
```

```python
# Sample historical monthly data
months = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
past_expenses = np.array([12000, 13000, 14000, 15000, 16000])

model = LinearRegression()
model.fit(months, past_expenses)

next_month = np.array([[6]])
predicted_expense = model.predict(next_month)[0]

print("\nPredicted Expense for Next Month:", int(predicted_expense))
```
